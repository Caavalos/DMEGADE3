```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.plot(2,2, 'x')
plt.plot(5,4,'o')
plt.plot(4,3,'^')

plt.xlabel('x axis name',
          fontsize = 18)
plt.xticks(fontsize = 15)

plt.ylabel('y axis name',
           fontsize = 18)
plt.yticks(fontsize = 15)

plt.title('A title', fontsize = 18)

#plt markers on a plane type
#the rest its for the axis size, name and title
```

```python
data = {'Names' : ['Name1', 'Name2', 'Name3', 'Name4'],
        'Age' : [33,42,41,52],
        'Weight' : [50,65,65,45]
       }

df = pd.DataFrame(data)
df
```

```python
line_plot = df.plot(kind = 'line')

plt.xlabel('indices',
          fontsize = 18)
plt.xticks(fontsize = 15)

plt.ylabel('justnumbers',
           fontsize = 18)
plt.yticks(fontsize = 15)

plt.title('A title', fontsize = 18)
plt.show()
```

```python
import seaborn as sns
```

```python
%ls datasets/
```

```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head()
```

```python
sns.lmplot (x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )
plt.title('Scatter plot of CWDistance vs Wingspan')
plt.show()
```

```python
sns.swarmplot( x= 'Gender',
              y = 'CWDistance',
              data = df
             )
```

```python
sns.boxplot ( data = df.loc[ :, ['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']
                           ]
            )

plt.xticks( fontsize = 17,
           rotation = 45)
plt.show(block = True)
```

```python
sns.boxplot ( data = df.loc[df['Gender']=='F', ['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']
                           ]
            )

plt.xticks( fontsize = 17,
           rotation = 45)
plt.title('Boxplots of Gender F')
plt.show(block = True)
```
